```python
import pandas as pd

url = 'https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv'

dados_covid = pd.read_csv(url)

dados_filtrados = dados_covid.iloc[:, 4:]
dados_filtrados.columns = pd.to_datetime(dados_filtrados.columns)
dados_filtrados = dados_filtrados.loc[:, (dados_filtrados.columns.year >= 2017) & (dados_filtrados.columns.year <= 2021)]

media_mortes_por_pais_ano = dados_filtrados.mean().reset_index()
media_mortes_por_pais_ano.columns = ['Data', 'Média de Mortes']
media_mortes_por_pais_ano[['País/Região', 'Província/Estado']] = dados_covid[['Country/Region', 'Province/State']]

media_mortes_por_pais_ano = media_mortes_por_pais_ano.sort_values(['País/Região', 'Data'])
media_mortes_por_pais_ano = media_mortes_por_pais_ano.drop_duplicates(subset=['País/Região'])

paises_sul_americanos = ['Argentina', 'Bolivia', 'Brazil', 'Chile', 'Colombia', 'Ecuador', 'Guyana', 'Paraguay', 'Peru', 'Suriname', 'Uruguay', 'Venezuela']
tabela = media_mortes_por_pais_ano[media_mortes_por_pais_ano['País/Região'].isin(paises_sul_americanos)]
tabela = tabela[['Data', 'País/Região', 'Média de Mortes']]
tabela.columns = ['Ano', 'País/Região', 'Número de Mortes']
tabela.reset_index(drop=True, inplace=True)

print(tabela)
```
